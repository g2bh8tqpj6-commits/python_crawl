練習題：計算體重指數 BMI
BMI（Body Mass Index，身體質量指數）是評估體重是否健康的常用指標。

計算公式：
BMI = 體重(公斤) ÷ 身高(公尺)²

題目要求：

建立變數 weight 儲存體重 72 公斤
建立變數 height 儲存身高 175 公分（注意：公式需要公尺，請先除以 100）
使用 ** 運算子計算身高的平方
計算 BMI 值並儲存在變數 bmi 中
使用 round() 函式將 BMI 四捨五入到小數點後 2 位
印出結果
參考輸出：
23.35

提示：

身高需轉換成公尺：height = 175 / 100
平方使用 ** 運算子：height ** 2
四捨五入：round(bmi, 2)

```python
weight = 72
height = 175 / 100

BMI = weight / (height **2)
BMI = round(BMI,2)
print(BMI)
```

練習題 1：計算飲料熱量

珍奶熱量計算方式：
總熱量 = 基底熱量 + (每份糖量 × 甜度份量)

條件如下：

珍奶基底熱量：240 大卡

每份糖量：30 大卡

甜度：使用者選擇 70% 甜（= 0.7 份）
題目要求：

建立變數

base 儲存基底熱量（240）

sugar_unit 儲存每份糖熱量（30）

sweetness 儲存甜度（70% → 0.7）

計算

使用 * 計算糖量熱量

使用 + 計算總熱量

使用 round() 四捨五入到小數後 1 位

存到變數 total_cal

```python
base= 240
sugar_unit = 30
sweetness = 70 / 100
總熱量 = base + (sugar_unit * sweetness )
總熱量 = round(總熱量,1)
print("飲料總熱量",總熱量)
```
